### 여러개 그래프 그리기 

```python
import numpy as np
```

```python
t=np.arange(0,5,0.2)
```

```python
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis([0,5,0,20])
```

### 그래프 영역채우기

```python
x=[1,2,3,4]
y=[1,4,9,16]

plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.fill_between(x[1:3],y[1:3],alpha=0.5)
plt.fill_betweenx(y[2:4],x[2:4],color='pink',alpha=0.5)

plt.show()
```

```python
x=[1,2,3,4]
y1=[1,4,9,16]
y2=[1,2,4,8]

plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('X')
plt.ylabel('Y')
plt.fill_between(x[1:3],y1[1:3],y2[1:3],color='gray',alpha=0.5)
plt.show()
```

### 그리드 설정하기

```python
a=np.arange(0,2,0.2)
plt.plot(a,a,'bo')
plt.plot(a,a**2,color='#e35f62',marker='*',linewidth=2)
plt.plot(a,a**3,color='springgreen',marker='^',markersize=9)
plt.grid(True,axis='y',color='red',alpha=0.4,linestyle='--')
plt.show()
```

### 눈금 표시하기

```python
a=np.arange(0,2,0.2)
plt.plot(a,a,'bo')
plt.plot(a,a**2,color='#e35f62',marker='*',linewidth=2)
plt.plot(a,a**3,color='springgreen',marker='^',markersize=9)
plt.xticks(a,labels=['Jan','','Feb','','Mar','','May','','June','','Jul',''])
plt.yticks(np.arange(0,7),('0','1GB','2GB','3GB','4GB','5GB','6GB'))

plt.tick_params(axis='x',direction='inout',length=3,pad=6,labelsize=13,labelcolor='green')
plt.tick_params(axis='y',direction='inout',length=3,pad=6,labelsize=13,labelcolor='green')
plt.show()
```

### 타이틀 지정하기

```python
a=np.arange(0,2,0.2)
plt.plot(a,a,'bo')
plt.plot(a,a**2,color='#e35f62',marker='*',linewidth=2)
plt.plot(a,a**3,color='springgreen',marker='^',markersize=9)
plt.xticks(a,labels=['Jan','','Feb','','Mar','','May','','June','','Jul',''])
plt.yticks(np.arange(0,7),('0','1GB','2GB','3GB','4GB','5GB','6GB'))
plt.title('Sample',loc='right',pad=25)
title_font={
    'fontsize':16,
    'fontweight':'bold'
}
plt.title('Sample',fontdict=title_font,loc='left',pad=25)
plt.show()
```

### 막대 그래프 그리기

```python
x=np.arange(3)
yaers=['2017','2018','2019']
values=[100,400,900]
plt.bar(x,values,width=0.4,align='edge',color='green')
plt.xticks(x,yaers)
plt.show()
```

### 수평 막대 그리기

```python
y=np.arange(3)
years=['2017','2018','2019']
values=[100,400,900]

plt.barh(y,values, align='edge',color='springgreen',
        edgecolor='gray',linewidth=3, tick_label=years,log=False)
plt.show()
```

### 산점도 그리기

```python
np.random.seed(1234)
N=50
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
area=(30*np.random.rand(N))**2

plt.scatter(x,y,s=area,c=colors,alpha=0.5)
plt.show()
```

### 히스토그램 그리기

```python
weight=[60,81,64,56,78,74,61,77,66,68,59,71,80,59,67,81,69,73,69,74,70,65]

plt.hist(weight,bins=4)
plt.show()
```

### 파이차트 그리기

```python
ratio=[34,32,16,18]
labels=['Apple','Banana','Melon','Grapes']

plt.pie(ratio,labels=labels,autopct='%.1f%%')
plt.show()
```

```python
ratio=[34,32,16,18]
labels=['Apple','Banana','Melon','Grapes']
colors=['#ff9999','#ffc000','#8fd9b6','#d395d0']
wedgeprops={'width':0.7,'edgecolor':'w','linewidth':5}

plt.pie(ratio,labels=labels,autopct='%.1f%%',startangle=260, counterclock=False,colors=colors,wedgeprops=wedgeprops)
plt.show()
```
